```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

**Task 4.Data Cleaning**

```python
#removing empty value rows
Data.dropna(inplace=True)
Data.isna().sum()

Data.duplicated().sum()

#deliting all duplucates
Data.drop_duplicates(inplace=True)
```

```python
# Changing Data type
Data.dtypes
Data['id'] = Data['id'].astype(object)
```

## EDA
**Task 5.Data Analysis**

**univeriate Analysis**

```python
# identifying uotlayers
sns.boxplot(data=Data,x='price')
plt.show()

Data2 = Data[Data['price'] <= 1500]
sns.boxplot(data=Data2,x='price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=Data2,x='price',kde=True,binwidth=20,bins=100,color='green')
plt.ylabel('Frequancy')
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=Data2,x='availability_365',)
plt.ylabel('Frequancy')
plt.title('availability_365 Distribution')
plt.show()
```

```python
# Average price for bed in each city
Data2['Bed_price'] = Data2['price']/Data2['beds'] 
Data2.groupby(by=Data2['neighbourhood_group'])['Bed_price'].mean()
sns.barplot(data=Data2,x='neighbourhood_group',y='Bed_price',hue='roo',color='red',saturation=0.70)
plt.show()
```

**Bivarint Analysis:**

Anlysing one column relationship with anethore column

one variable dependency on onether column

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=Data2,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=Data2,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
plt.show()
plt.savefig('pairplo.jpg',dpi=300)
```

```python
#Geografical distribution of AirBnB Listing
plt.figure(figsize=(8,6))
sns.scatterplot(data=Data2,x='longitude',y='latitude',hue='room_type')
plt.show()
```

```python
# heatmap - corelation of one variable with others for numerical columns
corelation = Data2[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corelation
plt.figure(figsize=(10))
sns.heatmap(data=corelation,annot=True)
plt.show()
plt.savefig('heatmap.jpg',dpi=300)
```
